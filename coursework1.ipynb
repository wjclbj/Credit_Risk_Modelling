```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dt  = pd.read_csv('422d3ec362a12646.csv', encoding='utf-8')
dt.head()
```

```python
dt_truncted = dt.query('conm == "S&P 500 Comp-Ltd"')
dt_truncted['datadate'] = pd.to_datetime(dt_truncted['datadate'], format = '%Y%m%d')
dt_truncted['pd'] = dt_truncted['prccm'] / dt_truncted['dvpsxm']
dt_truncted['div_g'] = dt_truncted['dvpsxm'].pct_change()
dt_truncted.head()
```

```python
dt_truncted['pd'].describe()
```

```python
dt_truncted['div_g'].describe()
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=2, sharex='col')
ax1.plot('datadate', 'pd', data=dt_truncted)
ax2.plot('datadate', 'div_g', data=dt_truncted)
plt.rcParams["figure.figsize"] = [32,18]
```
